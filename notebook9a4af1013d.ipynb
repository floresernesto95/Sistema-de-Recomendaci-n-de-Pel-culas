# **Sistemas de Recomendación**

Los sistemas de recomendación son herramientas y técnicas de software que proporcionan sugerencias de elementos a los usuarios. Estas sugerencias se relacionan con varios procesos de toma de decisiones, como qué productos comprar, qué música escuchar o qué noticias leer.

## **Tipos principales de sistemas de recomendación**

- ### **Filtrado Colaborativo**

    Este método se basa en la idea de que los usuarios similares tienen preferencias similares. Utiliza el historial de calificaciones o comportamientos de los usuarios para hacer predicciones.

- ### **Basado en Contenido**

    Este enfoque utiliza las características de los elementos para recomendar artículos adicionales con propiedades similares.

- ### **Sistemas Híbridos**

    Combinan múltiples técnicas para obtener mejores resultados.

## **Beneficios de los sistemas de recomendación**

Los sistemas de recomendación ofrecen numerosos beneficios tanto para los usuarios como para las empresas:

# **Fusión de DataFrames en Pandas**

```python
movies = movies.merge(credits, left_on='title', right_on='title')
```

#### **Fusión de DataFrames**
Utiliza el método merge() de Pandas para combinar los DataFrames 'movies' y 'credits'.

#### **Parámetros de merge**
- left_on='title': Especifica la columna del DataFrame izquierdo para la fusión.
- right_on='title': Especifica la columna del DataFrame derecho para la fusión.
- how: Determina el tipo de join.

  - 'inner' (predeterminado): Retorna solo las filas con coincidencias en ambos DataFrames.
  - 'left': Retorna todas las filas del DataFrame izquierdo y las coincidentes del derecho.
  - 'right': Retorna todas las filas del DataFrame derecho y las coincidentes del izquierdo.
  - 'outer': Retorna todas las filas cuando hay una coincidencia en cualquiera de los DataFrames.

#### **Cómo funciona**
- Pandas busca coincidencias entre los valores de la columna 'title' en ambos DataFrames.
- El tipo de join (especificado por 'how') determina qué filas se incluyen en el resultado final.
- Las filas se combinan basándose en la coincidencia de 'title' y el tipo de join seleccionado.

#### **Notas adicionales**
- Si no se especifica 'how', se realiza un inner join por defecto.
- La elección del tipo de join afecta significativamente el resultado, especialmente si hay datos no coincidentes.
- Es importante verificar que no se hayan perdido datos inesperadamente en el proceso de fusión.
- Considerar usar 'validate' para asegurar la integridad de la fusión (por ejemplo, 'one_to_one', 'one_to_many', etc.).

# **Preprocesamiento de Datos**

```python
import ast

movies['genres'] = movies['genres'].apply(lambda x: [i['name'] for i in ast.literal_eval(x)])
movies['keywords'] = movies['keywords'].apply(lambda x: [i['name'] for i in ast.literal_eval(x)])
movies['cast'] = movies['cast'].apply(lambda x: [i['name'] for i in ast.literal_eval(x)[:3]])
movies['crew'] = movies['crew'].apply(lambda x: [i['name'] for i in ast.literal_eval(x) if i['job'] == 'Director'])
```

#### **Transformaciones realizadas**

- Géneros y Palabras Clave:
   - Para 'genres' y 'keywords', se aplica la misma transformación.
   - Cada elemento se convierte de una cadena a una lista de Python usando `ast.literal_eval()`.
   - Se extrae el valor de la clave 'name' de cada diccionario en la lista.<br><br>

- Elenco (Cast):
   - Similar a géneros y palabras clave, pero solo se consideran los primeros tres elementos de la lista.
   - Esto probablemente se hace para limitar la información a los actores principales.<br><br>

- Equipo (Crew):
   - Se aplica una transformación más compleja.
   - Además de convertir la cadena a una lista y extraer los nombres, se filtra para incluir solo a los miembros del equipo cuyo trabajo ('job') es 'Director'.

#### **Propósito**

Este procesamiento tiene varios objetivos:

- Normalización de datos: Convierte cadenas de texto que representan listas y diccionarios en estructuras de datos de Python reales.
- Extracción de información relevante: Se enfoca en los nombres de géneros, palabras clave, actores y directores.
- Simplificación: Reduce la complejidad de los datos, especialmente para el elenco y el equipo, facilitando análisis posteriores.
- Preparación para análisis: Estructura los datos de una manera que sea más fácil de utilizar en algoritmos de recomendación o análisis de películas.

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies['tags'] = movies['tags'].apply(lambda x: ' '.join(x))
movies['tags'] = movies['tags'].apply(lambda x: x.lower())
```

#### **Operaciones**

- Creación de tags: Combina 'genres', 'keywords', 'cast' y 'crew' en una nueva columna 'tags'.
- Unión de elementos: Convierte la lista de tags en una sola cadena de texto.
- Conversión a minúsculas: Transforma todos los tags a minúsculas.

#### **Propósito**

- Consolida información de diferentes aspectos de cada película.
- Prepara el texto para análisis y búsquedas.
- Normaliza los datos para mejorar la precisión en análisis posteriores.

```python
movies = movies[['movie_id', 'title', 'overview', 'tags']]
```

# **Vectorización TF-IDF de tags de películas**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies['tags'])
```

Este código utiliza la técnica de vectorización TF-IDF (Term Frequency-Inverse Document Frequency) para convertir los tags de las películas en vectores numéricos permitiendo comparar y analizar películas basándose en la importancia relativa de los términos en sus tags, facilitando la identificación de similitudes y patrones en el conjunto de datos.

#### **Operaciones realizadas**

- Creación del vectorizador:
   - Se crea una instancia de `TfidfVectorizer` con el parámetro `stop_words='english'`.
   - Esto configura el vectorizador para eliminar las palabras comunes en inglés (stop words) durante el proceso.<br><br>

- Aplicación del vectorizador:
   - Se aplica el método `fit_transform()` del vectorizador a la columna 'tags' del DataFrame `movies`.
   - Este método realiza dos operaciones: primero, aprende el vocabulario del corpus (fit), y luego transforma los datos en una matriz TF-IDF (transform).<br><br>

- Resultado:
   - El resultado se almacena en `tfidf_matrix`, que es una matriz dispersa donde cada fila representa una película y cada columna representa un término único en el vocabulario.

#### **Propósito y significado**

- Representación numérica: Convierte el texto de los tags en una representación numérica que los algoritmos de aprendizaje automático pueden procesar.

- Ponderación de términos: TF-IDF pondera la importancia de cada término:
   - Términos frecuentes en una película pero raros en el conjunto de datos reciben puntuaciones altas.
   - Términos comunes en muchas películas reciben puntuaciones más bajas.<br><br>

- Eliminación de ruido: Al eliminar las stop words, se reduce el ruido en los datos, enfocándose en términos más significativos.

- Preparación para análisis: Esta matriz puede usarse para diversos análisis, como:
   - Cálculo de similitud entre películas.
   - Agrupación de películas similares.
   - Construcción de sistemas de recomendación basados en contenido.

#### **Otras técnicas de vectorización**

Además de TF-IDF, existen otras técnicas de vectorización comúnmente utilizadas en el procesamiento de lenguaje natural:

- Codificación One-Hot:
   - Representa cada palabra como un vector binario del tamaño del vocabulario.
   - Cada posición corresponde a una palabra, con un 1 si la palabra está presente y 0 si no.
   - Ventaja: Simple y fácil de entender.
   - Desventaja: Produce vectores muy dispersos y no captura la importancia relativa de las palabras.<br><br>

- Bag of Words (BoW):
   - Similar a One-Hot, pero cuenta la frecuencia de cada palabra en el documento.
   - Ventaja: Captura información sobre la frecuencia de las palabras.
   - Desventaja: No considera el orden de las palabras ni su importancia en el contexto más amplio.<br><br>

- Word Embeddings (ej. Word2Vec, GloVe):
   - Representa palabras como vectores densos de baja dimensión en un espacio continuo.
   - Captura relaciones semánticas entre palabras.
   - Ventaja: Preserva relaciones semánticas y contextuales entre palabras.
   - Desventaja: Requiere un gran corpus de texto para el entrenamiento y puede ser computacionalmente intensivo.<br><br>

- Doc2Vec:
   - Extensión de Word2Vec que genera embeddings para documentos enteros.
   - Útil cuando se necesita una representación del documento completo.<br><br>

- Transformadores (ej. BERT, GPT):
   - Modelos de lenguaje pre-entrenados que generan representaciones contextuales de palabras y frases.
   - Capturan el contexto bidireccional y producen embeddings dinámicos.
   - Ventaja: Estado del arte en muchas tareas de NLP.
   - Desventaja: Computacionalmente intensivos y requieren más recursos.

#### **Comparación con TF-IDF**
- TF-IDF ofrece un buen equilibrio entre simplicidad y efectividad.
- Captura la importancia de las palabras en el contexto del corpus, a diferencia de BoW o One-Hot.
- Es menos complejo y requiere menos recursos que Word Embeddings o Transformadores.
- Sin embargo, no captura relaciones semánticas profundas como lo hacen las técnicas más avanzadas.

La elección de la técnica de vectorización depende del problema específico, los recursos disponibles y el equilibrio deseado entre simplicidad y rendimiento.

# **Cálculo de Similitud del Coseno entre Películas**

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

#### **Operación realizada**

Se calcula la similitud del coseno entre todas las películas utilizando la matriz TF-IDF (`tfidf_matrix`) previamente calculada:

- La función `cosine_similarity` toma como argumentos la matriz TF-IDF dos veces, lo que resulta en el cálculo de similitud de cada película con todas las demás (incluyendo consigo misma).
- El resultado se almacena en la variable `cosine_sim`.

#### **Propósito y significado**

- Medida de similitud: La similitud del coseno mide la similitud entre dos vectores calculando el coseno del ángulo entre ellos. En este contexto, mide cuán similares son dos películas basándose en sus tags.

- Rango de valores: Los valores de similitud varían de -1 a 1, donde:
   - 1 indica vectores idénticos (películas muy similares)
   - 0 indica vectores ortogonales (películas sin relación)
   - -1 indica vectores opuestos (raramente ocurre en este contexto)<br><br>

- Matriz de similitud: `cosine_sim` es una matriz donde cada celda [i,j] representa la similitud entre la película i y la película j.

- Uso en recomendaciones: Esta matriz es fundamental para sistemas de recomendación basados en contenido, permitiendo encontrar rápidamente las películas más similares a una dada.

- Independencia de la magnitud: La similitud del coseno se basa en la dirección de los vectores, no en su magnitud, lo que la hace útil para comparar documentos de diferentes longitudes.

- Eficiencia computacional: Es computacionalmente eficiente, especialmente para matrices dispersas como las producidas por TF-IDF.

# **Función de Recomendación de Películas**

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = movies[movies['title'] == title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]
    return movies['title'].iloc[movie_indices]
```

Esta función, `get_recommendations`, implementa un sistema de recomendación de películas basado en la similitud del contenido.

#### **Parámetros de la función**

- `title`: El título de la película para la cual se buscan recomendaciones.
- `cosine_sim`: La matriz de similitud del coseno (por defecto, usa la matriz `cosine_sim` calculada previamente).

#### **Proceso de la función**

- Encontrar el índice de la película:
   - Localiza el índice de la película dada en el DataFrame `movies`.<br><br>

- Obtener puntuaciones de similitud:
   - Extrae las puntuaciones de similitud para la película dada de la matriz `cosine_sim`.
   - Crea una lista de tuplas (índice, puntuación) usando `enumerate`.<br><br>

- Ordenar las puntuaciones:
   - Ordena la lista de puntuaciones de mayor a menor similitud.<br><br>

- Seleccionar las mejores coincidencias:
   - Toma las 10 películas más similares (excluyendo la primera, que es la película misma).<br><br>

- Extraer índices de películas:
   - Obtiene los índices de las películas recomendadas.<br><br>

- Retornar títulos:
   - Devuelve los títulos de las películas recomendadas usando los índices obtenidos.

```python
print(get_recommendations('The Dark Knight Rises'))
```

   - Se utiliza `pickle.dump()` para serializar y guardar los datos.
   - Se guardan dos objetos: el DataFrame `movies` y la matriz de similitud `cosine_sim`.

#### **Propósito y beneficios**

- Persistencia de datos: 
   - Permite guardar los datos procesados para su uso posterior sin necesidad de reprocesar.<br><br>

- Eficiencia: 
   - Evita la necesidad de recalcular la matriz de similitud, que puede ser computacionalmente costosa.<br><br>

- Portabilidad: 
   - Los datos serializados pueden ser fácilmente transferidos y cargados en diferentes entornos.<br><br>

- Ahorro de tiempo: 
   - Reduce significativamente el tiempo de inicio del sistema de recomendación en futuras ejecuciones.<br><br>

- Consistencia: 
   - Asegura que se utilice el mismo conjunto de datos y cálculos en diferentes sesiones o despliegues.
